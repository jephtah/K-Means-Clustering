## KMeans Cluster Creation

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
new_data = (data.drop("Private", axis=1))
```

```python
new_data.head()
```

```python
y = kmeans.fit(new_data)
```

```python
y.labels_
```

```python
new_data
```

```python
kmeans.cluster_centers_
```

```python
def converter(cluster):
    if cluster=='Yes':
        return 1
    else:
        return 0
```

```python
data["Cluster"] = y.labels_
```

```python
data[["Cluster", "Private"]]
```

```python
def converter(privpub):
    if privpub =='Yes':
        return 0
    else:
        return 1
```

```python
data['real_clust'] = data['Private'].apply(converter)
```

```python
data[["Cluster", "real_clust"]]
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(classification_report(data['real_clust'],data["Cluster"]))
```

```python
print(confusion_matrix(data['real_clust'],data["Cluster"]))
```
